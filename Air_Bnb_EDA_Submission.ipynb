#### Chart - 1

````python
# Price Variation by Neighborhood Group

plt.figure(figsize=(12, 6))
scatter = plt.scatter(df['neighbourhood_group'], df['price'], alpha=0.5, c=df['price'], cmap='viridis')
plt.colorbar(scatter, label='Price')
plt.title('Price Variation by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
````

* Manhattan: Shows the highest price variation, with data points spread across nearly the entire price range.
* Brooklyn: Also has a wide variation in prices but fewer high-priced listings compared to Manhattan.
* Queens and Staten Island: Display more moderate price variations, concentrated towards the lower end of the price spectrum.
* Bronx: Has the least variation in prices, with most data points clustered at the bottom, indicating lower prices.

#### Chart - 4

````python
#Trend of Average Listing Price Over Time

# Convert 'last_review' to datetime format
df['last_review'] = pd.to_datetime(df['last_review'])

# Extract year and month from 'last_review'
df['year_month'] = df['last_review'].dt.to_period('M')

# Calculate average price by year and month
avg_price_over_time = df.groupby('year_month')['price'].mean()

# Plot the trend of average price over time
avg_price_over_time.plot(kind='line', marker='o')
plt.title('Trend of Average Listing Price Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
````

##### 1. Why did you pick the specific chart?

A line chart is chosen because it effectively shows trends over time, making it easy to observe any increases or decreases in average listing prices.

##### 2. What is/are the insight(s) found from the chart?

* Relative Stability (2017-2019): From 2017 onwards, the average listing price appears to stabilize within a narrower band, mostly hovering between 100  and
 200
* Lack of Clear Trend (Later Years): While there's a stabilization in the later years, there isn't a clear upward or downward trend during that period.

##### 3. Will the gained insights help creating a positive business impact?

````python
# Top 10 Hosts by Number of Listings
# Group by 'host_id' and calculate the number of listings for each host
top_hosts_listings = df.groupby('host_id')['id'].count()

# Sort the hosts by number of listings in descending order and get the top 10
top_10_hosts_listings = top_hosts_listings.sort_values(ascending=False).head(10)

# Create a DataFrame for plotting
top_10_hosts_df = top_10_hosts_listings.reset_index()
top_10_hosts_df.columns = ['host_id', 'number_of_listings']

# Merge with the original dataframe to get host names
top_10_hosts_df = top_10_hosts_df.merge(df[['host_id', 'host_name']].drop_duplicates(), on='host_id', how='left')

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='number_of_listings', y='host_name', data=top_10_hosts_df, palette='viridis',hue='host_name', legend=False)
plt.title('Top 10 Hosts by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Host Name')
plt.show()
````

````python
# Top 5 Hosts with Highest Availability and Lowest Availability

# Find the top 5 hosts with the highest and lowest availability
top_5_highest_availability = df.nlargest(5, 'availability_365')
top_5_lowest_availability = df.nsmallest(5, 'availability_365')

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot for top 5 highest availability
sns.barplot(ax=axes[0], x=top_5_highest_availability['host_name'], y=top_5_highest_availability['availability_365'], color='skyblue')
axes[0].set_title('Top 5 Hosts with Highest Availability')
axes[0].set_xlabel('Host Name')
axes[0].set_ylabel('Availability (days)')
axes[0].tick_params(axis='x', rotation=45)

# Plot for top 5 lowest availability
sns.barplot(ax=axes[1], x=top_5_lowest_availability['host_name'], y=top_5_lowest_availability['availability_365'], color='orange')
axes[1].set_title('Top 5 Hosts with Lowest Availability')
axes[1].set_xlabel('Host Name')
axes[1].set_ylabel('Availability (days)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````
